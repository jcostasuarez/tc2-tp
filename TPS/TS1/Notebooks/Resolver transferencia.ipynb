```python
from sympy import simplify
from sympy.physics.control.lti import TransferFunction
from sympy.physics.control.control_plots import pole_zero_plot, bode_plot
from sympy import symbols


#Acá tenes que declarar las variables
s = symbols('s')

K = 0.2512
expr1 = K*(s**2 + 1/K) / (s ** 2 + s * 1.414 + 1)

display(simplify(expr1))

tf1 = TransferFunction.from_rational_expression(expr1,var =s)

bode_plot(tf1)

```

```python

expr1 = (s**2 + s * 0.707 + 1) / (s ** 2 + s * 1.414 + 1)

display(simplify(expr1))

tf1 = TransferFunction.from_rational_expression(expr1,var =s)

bode_plot(tf1)
```

```python

expr1 = (s**2 - s * 1.414 + 1) / (s ** 2 + s * 1.414 + 1)

display(simplify(expr1))

tf1 = TransferFunction.from_rational_expression(expr1,var =s)

bode_plot(tf1)
```
