```python
### SIMULACIÓN NUMÉRCIA###

import numpy as np
from scipy import signal as sig
from splane import analyze_sys, pretty_print_bicuad_omegayq

num = np.array([1,-1000]) 
den = np.array([1,1000])
    
mi_sos = sig.TransferFunction(num,den)
analyze_sys(mi_sos)
```

```python
### SIMULACIÓN SIMBÓLICA###

from sympy.abc import s
from sympy.physics.control.lti import TransferFunction
from sympy.physics.control.control_plots import pole_zero_plot, bode_plot

R1 = 1
R2 = 1
R3 = 1000
C = 1E-6

tf1 = TransferFunction(s - R2/(C*R1*R3), s +1/(C*R3), s)
pole_zero_plot(tf1)
bode_plot(tf1, initial_exp=-1, final_exp=6)
```
