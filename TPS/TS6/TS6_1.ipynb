# Prototipo Pasa-Bajos


```python
from sympy import simplify
from sympy.physics.control.lti import TransferFunction
from sympy.physics.control.control_plots import pole_zero_plot, bode_plot
from sympy import symbols
import matplotlib.pyplot as plt


#Acá tenes que declarar las variables
s,w0 = symbols('s w0')

expr = 1/9*(s**2 + 9) / ((s ** 2 + s+ 1)*(s+1))

display(simplify(expr))

tf1 = TransferFunction.from_rational_expression(expr,var =s)

bode_plot(tf1,initial_exp=-1, final_exp=2)
```

# Expresión de H(s) del filtro pasa-altos normalizado      

```python
expr1 = expr.subs(s, 1/s)

expr1 = simplify(expr1)

display(expr1)

tf2 = TransferFunction.from_rational_expression(expr1,var =s)

bode_plot(tf2, initial_exp=-2, final_exp=2)

pole_zero_plot(tf2)
```

```python
f0 = 300

expr2 = expr1.subs(s, s/f0)

expr2 = simplify(expr2)

display(expr2)

tf3 = TransferFunction.from_rational_expression(expr2,var =s)

bode_plot(tf3, initial_exp=1, final_exp=3)
```
