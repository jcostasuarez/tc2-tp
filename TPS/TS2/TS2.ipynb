## Akerberg-Mossberg Low Pass

$$- \frac{G_{1}}{\frac{C^{2} G_{5} s^{2}}{G_{3} G_{4}} + \frac{C G_{2} G_{5} s}{G_{3} G_{4}} + G_{3}} $$

$$- \frac{10}{s^{2} + 0.2 s + 1}
$$

```python
from sympy import symbols, cancel, pprint, latex
from sympy.solvers.solveset import linsolve
from sympy.physics.control.lti import TransferFunction
from sympy.physics.control.control_plots import pole_zero_plot, bode_plot


G1, G2, G3, G4, G5, C, s = symbols('G1 G2 G3 G4 G5 C s')

Equation = - G1 / (G5 * s**2 * C**2 / (G3 * G4) + G5 * G2 * s * C / (G3 * G4) + G3)

print("Latex equation: ")
print(latex(Equation))

Equation = Equation.subs([(G1, 10),(G2, 1/5),(G3, 1),(G4, 1),(G5, 1),(C, 1)])

Transf = TransferFunction.from_rational_expression(Equation,var =s)

Transf = cancel(Transf)

print()
print()
print("Transference(H(s)): ")
pprint(Transf) # I

print()
print()
pole_zero_plot(Transf)

print()
print()
bode_plot(Transf)  
```

## Akerberg-Mossberg Band Pass
$$- \frac{G_{1} s}{C s^{2} + G_{2} s + \frac{G_{3}^{2} G_{4}}{C G_{5}}}$$

```python
from sympy import symbols, cancel, pprint
from sympy.solvers.solveset import linsolve
from sympy.physics.control.lti import TransferFunction
from sympy.physics.control.control_plots import pole_zero_plot, bode_plot


G1, G2, G3, G4, G5, C, s = symbols('G1 G2 G3 G4 G5 C s')

Equation = -G1 * s / (s**2 * C + G2 * s + (G3**2 * G4/( C * G5)) )


print("Latex equation: ")
print(latex(Equation))

Equation = Equation.subs([(G1, 10),(G2, 1/5),(G3, 1),(G4, 1),(G5, 1),(C, 1)])

Transf = TransferFunction.from_rational_expression(Equation,var =s)

Transf = cancel(Transf)

print()
print()
print("Transference(H(s)): ")
pprint(Transf) # I

print()
print()
pole_zero_plot(Transf)

print()
print()
bode_plot(Transf)  
```
